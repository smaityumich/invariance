```python
import json
import re
import pandas as pd
pd.set_option('display.max_rows', None)
with open('out.json', 'r') as f:
    result = f.read()
    
res = re.split(r'\n', result)
res.pop()
res_dict = dict()
for index, string in enumerate(res):
    res_dict[index] = eval(string)
    
df = pd.DataFrame(res_dict).T
```
