```python
import json
import re
import pandas as pd
pd.set_option('display.max_rows', None)
with open('out_iter.json', 'r') as f:
    result = f.read()
    
result
res = re.split(r'\n', result)
res
```

```python
res_dict = dict()
for index, string in enumerate(res):
    if string == '':
        continue
    else:
        res_dict[index] = eval(string)
    
df = pd.DataFrame(res_dict).T
```
